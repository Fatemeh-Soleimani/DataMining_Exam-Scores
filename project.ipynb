```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('exams.csv')
```

```python
print(data.head())
print(data.info())
print(data.describe())

```

```python
# Distribution of scores
sns.histplot(data['math score'], kde=True, label='Math')
sns.histplot(data['reading score'], kde=True, label='Reading')
sns.histplot(data['writing score'], kde=True, label='Writing')
plt.title('Distribution of Scores')
plt.xlabel('Score')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
# Pair plot of scores and factors
sns.pairplot(data, vars=['math score', 'reading score', 'writing score'], hue='test preparation')
plt.show()

# Box plot of scores and factors
for col in ['gender', 'race/ethnicity', 'parental level', 'lunch', 'test preparation course
']:
    sns.boxplot(x=col, y='math score', data=data)
    plt.title(col)
    plt.show()
```

```python
# Distribution of scores by test preparation
sns.histplot(data[data['test preparation course'] == 'completed']['math score'], kde=True, label='Completed')
sns.histplot(data[data['test preparation course'] == 'none']['math score'], kde=True, label='None')
plt.title('Distribution of Math Scores by Test Preparation')
plt.xlabel('Score')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
sns.histplot(data[data['test preparation course'] == 'completed']['reading score'], kde=True, label='Completed')
sns.histplot(data[data['test preparation course'] == 'none']['reading score'], kde=True, label='None')
plt.title('Distribution of Reading Scores by Test Preparation')
plt.xlabel('Score')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
sns.histplot(data[data['test preparation course'] == 'completed']['writing score'], kde=True, label='Completed')
sns.histplot(data[data['test preparation course'] == 'none']['writing score'], kde=True, label='None')
plt.title('Distribution of Writing Scores by Test Preparation')
plt.xlabel('Score')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
# Correlation matrix
corr_matrix = data.corr()
plt.figure(figsize=(8, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
