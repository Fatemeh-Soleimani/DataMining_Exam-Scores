```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# data

```python
data = pd.read_csv('exams.csv')
```

```python
print(data.head())
```

```python
print(data.info())
```

```python
print(data.describe())
```

# Distribution of scores

```python

sns.histplot(data['math score'], kde=True, label='Math', color='blue')
sns.histplot(data['reading score'], kde=True, label='Reading', color='red')
sns.histplot(data['writing score'], kde=True, label='Writing', color='yellow')
plt.title('Distribution of Scores')
plt.xlabel('Score')
plt.ylabel('Count')
plt.legend()
plt.show()

```

# Box plot of scores and factors

```python
# Pair plot of scores and factors
# sns.pairplot(data, vars=['math score', 'reading score', 'writing score'], hue='test preparation course')
# plt.show()

for col in ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']:
    sns.boxplot(x=col, y='math score', data=data)
    plt.title(col)
    plt.show()
```
