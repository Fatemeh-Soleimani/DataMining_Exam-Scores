```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# data

```python
data = pd.read_csv('exams.csv')
```

```python
print(data.head())
```

```python
print(data.info())
```

```python
print(data.describe())
```

# Distribution of scores

```python

sns.histplot(data['math score'], kde=True, label='Math', color='blue')
sns.histplot(data['reading score'], kde=True, label='Reading', color='red')
sns.histplot(data['writing score'], kde=True, label='Writing', color='yellow')
plt.title('Distribution of Scores')
plt.xlabel('Score')
plt.ylabel('Count')
plt.legend()
plt.show()

```

# Box plot of scores and factors

### math score

```python
# Pair plot of scores and factors
# sns.pairplot(data, vars=['math score', 'reading score', 'writing score'], hue='test preparation course')
# plt.show()

for col in ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']:
    sns.boxplot(x=col, y='math score', data=data)
    plt.title(col)
    plt.show()
```

### writing score

```python
for col in ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']:
    sns.boxplot(x=col, y='writing score', data=data)
    plt.title(col)
    plt.show()
```


### reading score

```python
for col in ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']:
    sns.boxplot(x=col, y='reading score', data=data)
    plt.title(col)
    plt.show()
```

```python
# Distribution of scores by test preparation
sns.histplot(data[data['test preparation course'] == 'none']['math score'], kde=True, label='None',color='red')
sns.histplot(data[data['test preparation course'] == 'completed']['math score'], kde=True, label='Completed',color='green')

plt.title('Distribution of Math Scores by Test Preparation')
plt.xlabel('Score')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
sns.histplot(data[data['test preparation course'] == 'none']['reading score'], kde=True, label='None',color='red')
sns.histplot(data[data['test preparation course'] == 'completed']['reading score'], kde=True, label='Completed',color='green')

plt.title('Distribution of Reading Scores by Test Preparation')
plt.xlabel('Score')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
sns.histplot(data[data['test preparation course'] == 'none']['writing score'], kde=True, label='None',color='red')
sns.histplot(data[data['test preparation course'] == 'completed']['writing score'], kde=True, label='Completed',color='green')

plt.title('Distribution of Writing Scores by Test Preparation')
plt.xlabel('Score')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
# Correlation matrix
corr_matrix = data.corr()
plt.figure(figsize=(8, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
